## Question 1:

```python
import numpy as np
from scipy.optimize import minimize
```

```python
def obj(xy):
    # xy[0] is Labor
    # xy[1] is Capital
    
    mac = -(0.05*(xy[0]**(2/3))*xy[1]**(1/3)) 
    
    
    return mac
```

```python
best_loc = minimize(obj,[0,0],method='BFGS')
best_loc.x
```

```python
def confun(xy):
    return 12*xy[0]+15*xy[1] - 100000
```

```python
constr1 = {'type':'eq', 'fun': confun}
# then we put all our constraint dictionaries into a list
constraints = [constr1]
```

```python
best_loc_north = minimize(obj,[2000,5000],constraints=constraints)
```

```python
best_loc_north.x
```

```python
-best_loc_north.fun
```

## Question 4

```python
nfl = pd.read_csv('nflratings.csv', header=None, names = ['Week','HTI','VTI','HTS','VTS'])
```

```python
nfl
```

```python
nfl['APS'] = nfl['HTS'] - nfl['VTS']
```

```python
def obj(ratings):
    # xy[0] is Labor
    # xy[1] is Capital
    
    pred_spread = (ratings[nfl['HTI']] - ratings[nfl['VTI']] + ratings[32])
    pred_error = sum((pred_spread - nfl['APS'])**2)
    
    return pred_error
```

```python
a = [85]*32 + [3]
```

```python
best_loc = minimize(obj,a,method='BFGS')
best_loc.x
```

```python
best_loc.fun
```

```python
nfl['pred_spread'] = (best_loc.x[nfl['HTI']] - best_loc.x[nfl['VTI']] + best_loc.x[32])
```

```python
nfl
```

```python
nfl['a_win'] = np.where(nfl['APS'] > 0, 1,0)
nfl['p_win'] = np.where(nfl['pred_spread'] > 0, 1,0)
```

```python
nfl
```

```python
sum(nfl['a_win'] == nfl['p_win'])
```
