```python
def rankof(A):
 
    R = []
 
    for i in range(len(A)):
        r = 1
        for j in range(len(A)):
            if j != i and A[j] > A[i]:
                r = r+1     
    
        R.append(r) 
        
    return R

rankof(r)



#R = [0 for x in range(len(A))]
```
